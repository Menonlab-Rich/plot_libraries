```python
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.interpolate import interp1d
import numpy as np

x = [0, 2, 4, 6, 8, 10, 12, 20, 30, 40, 50, ]
y = [15, 19, 21, 22, 22, 22.5, 22.6, 22.9, 23, 23, 24, ]
f = interp1d(x, y, kind='slinear')
xnew = np.linspace(0, 50, 500)
# Plot the data using seaborn
sns.lineplot(x=xnew, y=f(xnew))
plt.xlabel('Excitation Laser Power (mW)')
plt.ylabel('Kilo-Events per second')
plt.title('Events vs power')
plt.show()
```
